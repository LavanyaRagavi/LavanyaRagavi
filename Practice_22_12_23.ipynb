```python
a = [11,24,32,45,48,50,54, 96]
def binary (a, n):
  low = 0
  high = len(a)-1
  mid = 0

  while low <= high:
      mid = (high + low)//2

      print(mid)
      if a[mid] < n:
          low = mid + 1
          print(low)

      elif a[mid] >n:
          high = mid - 1
      else:
          return mid

  return -1
```
